### 3.4. Match or not: What about common interests? ###

```python
dataset["match"].value_counts()/len(dataset)
```

To increase your chances of going to a second date, you must first get as many matches as possible during the event.  
It doesn't seem so obvious since only 16% of encounters lead to a match.  
  
In this part, we are going to try to understand the influence of the main centers of interest on the alchemy between two people.

```python
# Prepare the df_interests matrix for further analysis
to_keep = ["iid", "pid", "gender", "match", "sports", "tvsports", "exercise", "dining", "museums", "art", "hiking", "gaming", "clubbing", "reading", "tv", "theater", "movies", "concerts", "music", "shopping", "yoga"]
list_interests = ["sports", "tvsports", "exercise", "dining", "museums", "art", "hiking", "gaming", "clubbing", "reading", "tv", "theater", "movies", "concerts", "music", "shopping", "yoga"]
dataset_interests = dataset[to_keep].set_index(["iid", "pid"])

# Remove raws with missing values
ids_to_drop = dataset_interests[dataset_interests.isna().any(axis = 1)].index.get_level_values(0).unique().tolist()
df_interests = dataset_interests.drop(ids_to_drop, axis = 0, level = "iid").drop(ids_to_drop, axis = 0, level = "pid").reset_index()
df_interests = df_interests.dropna(axis = 0)

# Get subject's favorite interest
df_interests["subject's interest"] = df_interests[list_interests].idxmax(axis = 1)

# Get partner's favorite interest
Pint = []
for pid in df_interests["pid"]:
    Pint.append(df_interests.loc[df_interests["iid"] == pid, list_interests].iloc[0].idxmax())
df_interests["partner's interest"] = Pint

# Format final dataframe
df_favorite_interests = df_interests[["iid", "pid", "subject's interest", "partner's interest", "gender", "match"]].set_index(["iid", "pid"])

display(df_favorite_interests)
```

We will visualize in the form of a heatmap the possibility of obtaining a match according to the main interest of the two participants.  
Some activities might never lead to a match in this dataset, so we'll eliminate those combinations to make it easier to read the results and avoid the result to be mistaken with a 0% chance of getting a match.

```python
# Define interests that never lead to a match in the study
from itertools import product
to_drop = set([interests_pair[0]
                for interests_pair in product(list_interests, list_interests)
                if df_favorite_interests[(df_favorite_interests["subject's interest"] == interests_pair[0]) & (df_favorite_interests["partner's interest"] == interests_pair[1])].empty])
list_interests_clean = [interest for interest in list_interests if interest not in to_drop]

# Remove rows with these interests
mask1 = df_favorite_interests["partner's interest"].apply(lambda x : True if x not in to_drop else False) 
mask2 = df_favorite_interests["subject's interest"].apply(lambda x : True if x not in to_drop else False)
mask = mask1 & mask2
df_favorite_interests = df_favorite_interests[mask]

display(df_favorite_interests)
```

```python
interest_pairs_fig_gender = px.density_heatmap(df_favorite_interests, x = "subject's interest", y = "partner's interest", z = 'match', histfunc = "avg", 
                           category_orders = {"subject's interest" : list_interests_clean,
                                              "partner's interest" : list(reversed(list_interests_clean))},
                           marginal_x = 'histogram',
                           range_color = [0, df_favorite_interests.groupby(["subject's interest", "partner's interest"])['match'].mean().max()], 
                           title = "Chances of getting a match by favorite interest pairs", height = 600, width = 1000)

interest_pairs_fig_gender.show()
```

From the heatmap presented above, and knowing that the average match is 16% during this study, we can see that:  
- the people who seem to fit the most are hikers and tv sports fans (40% of matches during these meetings),
- tv sports fans also get along very well with people who like music, but never matched during the study with people who like museums or concerts,
- people who enjoy dining get along similarly with everyone except people who enjoy working out,
- movie fans generally don't get along with a lot of people (the frequency of matches being generally lower than the global average), and absolutely not with readers,
- sport enthusiasts, who represent more than 1/4 of the population, seem to get along with everyone, but still have a preference for hikers,
- some categories never match among themselves: art lovers, hickers, or surprisingly concert fans, but some groups being poorly represented could be an explanation for the lack of matches.
